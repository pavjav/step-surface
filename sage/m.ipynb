```python
def default_position(label):
    l = ZZ(label)
    if(l==1):
        return SG(0,0) #(b+c)x,(a+c)y
    if(l==2):
        return SG(1,0) #(b+c)x + c, (a+c)y
    if(l==3):
        return SG(1,1) #(b+c)x, (a+c)y + c
    if(l==4):
        return SG(2,1)
    if(l==5):
        return SG(2,2)
    if(l==6):
        return SG(3,2)
    if(l==7):
        return SG(3,3)
    if(l==8):
        return SG(4,3)
    if(l==9):
        return SG(4,4)
    if(l==10):
        return SG(5,4)
    if(l==11):
        return SG(5,5)
    if(l==12):
        return SG(6,5)
```

```python
def default_position_canon(label):
    l = ZZ(label)
    if(l==0):
        return SG(0,0) #(b+c)x,(a+c)y
    if(l==1):
        return SG(1,0) #(b+c)x + c, (a+c)y
    if(l==2):
        return SG(1,1) #(b+c)x, (a+c)y + c
    if(l==3):
        return SG(2,1)
    if(l==4):
        return SG(2,2)
    if(l==5):
        return SG(3,2)
    if(l==6):
        return SG(3,3)
    if(l==7):
        return SG(4,3)
    if(l==8):
        return SG(4,4)
    if(l==9):
        return SG(5,4)
    if(l==10):
        return SG(5,5)
    if(l==11):
        return SG(6,5)
```

```python
As=(s)
GS=As.graphical_surface(cached=False,edge_labels=False, default_position_function=default_position,polygon_labels=False)
v_x=49
v_y=31
unit=1
v_x/=unit
v_y/=unit
v=As.tangent_vector(1,(sqrt(2)/2,0),(v_x,v_y))
traj=v.straight_line_trajectory()
traj.flow(900)
print 'Trajectory is '
if(traj.is_closed()):
    print 'Closed'
else:
    print 'Drift'
gtraj=traj.graphical_trajectory(GS)
GS.make_all_visible()
p=[]
p.append(GS.plot())
p.append(gtraj.plot())
for i in range(1,13):
    vec=As.tangent_vector(i,(0.5,0.5),((i)%2,(i+1)%2))
    t=vec.straight_line_trajectory()
    t.flow(0.5)
    g=t.graphical_trajectory(GS)
    if i%2==1:
        p.append(g.plot(color="blue"))
    else:
        p.append(g.plot(color="red"))
GS.make_all_visible()
sum(p)
    
```
